```python
# Dependencies
#SQL Alchemy
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create engine and connection to database
engine = create_engine(f'postgresql://{Username}:{Password}@localhost:5433/employees')
connection = engine.connect()
```

```python
# Query to get data from employees table
employee_data = pd.read_sql("select * from employees", connection)
# Diplaye the employees table
employee_data.head()
```

```python
# Create a new table by joining the employees and salaries tables
employee_salary_data = pd.read_sql('select employees.emp_no, salaries.salary from employees as employees inner join salaries as salaries on employees.emp_no = salaries.emp_no', connection)
# Diplaye the new table
employee_salary_data.head()
```

```python
# Create a histogram to visualise the most common salary ranges for employees.
plt.figure(figsize=(10, 5))
plt.hist(employee_salary_data["salary"],histtype='bar',color='y', bins=15,edgecolor='k')
plt.ylabel(f'Number of Employees',fontsize=10)
plt.xlabel('Salary Range',fontsize=10)
plt.title("Employee Salary Distribution",fontsize=15)
plt.show()
```

```python
# Query to get data from titles table
title_data = pd.read_sql("select * from titles", connection)
# Diplaye data
title_data.head()
```

```python
# Rename the column to anable merging on that column
employee_data = employee_data.rename(columns={"emp_title_id":"title_id"})
employee_data.head()
```

```python
# Merge employee and titles tables
employee_title_data = pd.merge(employee_data, title_data, on="title_id", how="inner")
employee_title_data.head()
```

```python
# Merge tables
merged_data= pd.merge(employee_title_data, employee_salary_data, on="emp_no", how="inner")
merged_data.head()
```

```python
# Create a new dataframe that only contains required columns for this analysis
new_df=merged_data[['title','salary']]
new_df.head()
```

```python
# Create a bar chart of average salary by title
grouped_data=new_df.groupby(['title']).mean().plot(kind='bar',stacked=False)
plt.title("Average Employee Salary by Title")
plt.xlabel("Employee Titles")
plt.ylabel("Average Salary")
```
